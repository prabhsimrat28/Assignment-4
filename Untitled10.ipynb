```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

data = pd.read_csv("data.csv", encoding="latin1")


x = data["no2"].dropna().values

r = 102317135
a_r= 0.5*(r%7)
b_r= 0.3*((r%5)+1)

print("a_r =",a_r)
print("b_r =",b_r)

z=x+a_r*np.sin(b_r*x)


z =(z-np.mean(z))/np.std(z)
z = torch.tensor(z,dtype=torch.float32).view(-1,1).to(device)


```

```python
# -------------------------
# GAN MODEL
# -------------------------

class Gnet(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(1,32),
            nn.ReLU(),
            nn.Linear(32,64),
            nn.ReLU(),
            nn.Linear(64,1)
        )
    def forward(self,x):
        return self.net(x)

class Dnet(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(1,64),
            nn.LeakyReLU(0.2),
            nn.Linear(64,32),
            nn.LeakyReLU(0.2),
            nn.Linear(32,1),
            nn.Sigmoid()
        )
    def forward(self,x):
        return self.net(x)

G = Gnet().to(device)
D = Dnet().to(device)

criterion = nn.BCELoss()

g_opt = optim.Adam(G.parameters(),lr=0.0002)
d_opt = optim.Adam(D.parameters(),lr=0.0002)

```

```python
epochs = 3000
batch_size = 128

for epoch in range(epochs):

    idx = np.random.randint(0,len(z),batch_size)
    real = z[idx]

    # train discriminator
    noise = torch.randn(batch_size,1).to(device)
    fake = G(noise).detach()

    real_labels = torch.ones(batch_size,1).to(device)
    fake_labels = torch.zeros(batch_size,1).to(device)

    d_real = D(real)
    d_fake = D(fake)

    d_loss_real = criterion(d_real,real_labels)
    d_loss_fake = criterion(d_fake,fake_labels)

    d_loss = d_loss_real + d_loss_fake

    d_opt.zero_grad()
    d_loss.backward()
    d_opt.step()

    # train generator
    noise = torch.randn(batch_size,1).to(device)
    fake = G(noise)

    g_loss = criterion(D(fake),real_labels)

    g_opt.zero_grad()
    g_loss.backward()
    g_opt.step()

    if epoch % 500 == 0:
        print("epoch:",epoch,"d_loss:",d_loss.item(),"g_loss:",g_loss.item())

print("training done")
```

```python

noise = torch.randn(50000,1).to(device)
z_fake = G(noise).detach().cpu().numpy()

z_real = z.detach().cpu().numpy()


plt.figure(figsize=(10,6))


plt.hist(z_real,bins=100,density=True,alpha=0.4,label="real z")


plt.hist(z_fake,bins=100,density=True,alpha=0.4,label="fake z")


kde = KernelDensity(kernel='gaussian',bandwidth=0.2)
kde.fit(z_fake)

z_range = np.linspace(min(z_fake),max(z_fake),1000).reshape(-1,1)
log_density = kde.score_samples(z_range)

plt.plot(z_range,np.exp(log_density),linewidth=2,label="GAN pdf")

plt.legend()
plt.title("PDF learned by GAN")
plt.show()
```
